**Problema 1** 

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# geracao de forma de onda
f = 64000  # freq de amostragem para representação do sinal
T = 0.006     # duracao em segundos
t = np.arange(f*T)/f  # eixo do tempo
f1 = 3200  # feq1 do sinal original
f2 = 600   # feq2 do sinal original
f3 = 300   # feq3 do sinal original
#
#Sinal original:
x = np.cos(2*np.pi*f1*t) + 0.5*np.cos(2*np.pi*f2*t) + 0.01*np.cos(2*np.pi*f3*t)
plt.figure(figsize=(12,5))
plt.plot(t, x)
plt.title('Sinal original x(t) no tempo')
plt.xlabel('t, s')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

```
# frequência de amostragem de x(t)
fs = 6400
nfs = int(np.floor(f/fs)) # no. amostras em 1/fs
# sinal s amostrado a fs [Hz]
xs = x[::nfs]# seleciona os pontos a cada nfs
ts = t[::nfs]

plt.figure(figsize=(12,5))
plt.plot(t, x)
plt.title('x(t)')
plt.xlabel('t, s')
plt.grid()
# sinal discreto no tempo
plt.stem(ts, xs, 'k')
plt.plot(ts, xs, 'g:')
plt.title('Sinal amostrado a ' + str(fs) + ' Hz')
plt.show()
```

```
# sinal obtido da amostragem
plt.figure(figsize=(12,5))
n = range(len(xs))
plt.stem(n, xs, 'k')
plt.plot(n, xs, 'g:')
plt.title('Amostras do sinal no tempo')
plt.xlabel('n')
plt.grid()
plt.show()
```

```
# PARA FS' = FS/4
fs_ = fs/4
nfs = int(np.floor(f/fs_)) 
xs = x[::nfs]
ts = t[::nfs]
#
plt.figure(figsize=(12,5))
n = range(len(xs))
plt.stem(n, xs, 'k')
plt.plot(n, xs, 'g:')
plt.title('Amostras do sinal no tempo para FS/4')
plt.xlabel('n')
plt.grid()
plt.show()
```
